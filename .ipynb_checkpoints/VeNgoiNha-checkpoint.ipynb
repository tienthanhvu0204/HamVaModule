```python
import turtle
import my_module

screen = turtle.Screen ()
screen.setup (1200, 800)
screen.title ("Vẽ Ngôi Nhà")

turtle.penup ()
x0 = -400
y0 = -250
turtle.goto (x0, y0)
turtle.pendown ()

my_module.rectangle (150, 200, "#00BFFF")
turtle.fd (150)
my_module.rectangle (150, 200, "#00BFFF")
turtle.fd (150)
my_module.rectangle (200, 200, "#778899")
turtle.fd (200)
my_module.rectangle (150, 200, "#00BFFF")
turtle.fd (150)
my_module.rectangle (150, 200, "#00BFFF")
turtle.fd (150)

turtle.penup ()
turtle.goto (x0, y0 + 200)
turtle.pendown ()

my_module.rectangle (150, 200, "#00BFFF")
turtle.fd (150)
my_module.rectangle (150, 200, "#00BFFF")
turtle.fd (150)
my_module.rectangle (200, 200, "#778899")
turtle.fd (200)
my_module.rectangle (150, 200, "#00BFFF")
turtle.fd (150)
my_module.rectangle (150, 200, "#00BFFF")
turtle.fd (150)

turtle.penup ()
turtle.goto (x0 + 40, y0 + 70)
turtle.pendown ()

my_module.rectangle (40, 70, "white")
turtle.penup ()
turtle.fd (150)
turtle.pendown ()
my_module.rectangle (40, 70, "white")
turtle.penup ()
turtle.fd (350)
turtle.pendown ()
my_module.rectangle (40, 70, "white")
turtle.penup ()
turtle.fd (150)
turtle.pendown ()
my_module.rectangle (40, 70, "white")

turtle.penup ()
turtle.goto (x0 + 40, y0 + 270)
turtle.pendown ()

my_module.rectangle (40, 70, "white")
turtle.penup ()
turtle.fd (150)
turtle.pendown ()
my_module.rectangle (40, 70, "white")
turtle.penup ()
turtle.fd (150)
turtle.pendown ()
my_module.rectangle (40, 70, "white")
turtle.penup ()
turtle.fd (80)
turtle.pendown ()
my_module.rectangle (40, 70, "white")
turtle.penup ()
turtle.fd (120)
turtle.pendown ()
my_module.rectangle (40, 70, "white")
turtle.penup ()
turtle.fd (150)
turtle.pendown ()
my_module.rectangle (40, 70, "white")

turtle.penup ()
turtle.goto (x0 + 350, y0)
turtle.pendown ()

my_module.rectangle (50, 150, "#008080")
turtle.fd (50)
my_module.rectangle (50, 150, "#008080")

turtle.penup ()
turtle.goto (x0 - 50, y0 + 400)
turtle.pendown ()
turtle.color ("#A52A2A")
turtle.begin_fill ()
turtle.goto (x0 + 400, y0 + 500)
turtle.goto (x0 + 850, y0 + 400)
turtle.end_fill ()

turtle.hideturtle ()
turtle.done ()

```
